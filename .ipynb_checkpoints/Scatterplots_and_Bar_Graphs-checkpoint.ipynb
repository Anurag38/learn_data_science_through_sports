# Scatterplots

```python
import matplotlib
from matplotlib import pyplot as plt
import numpy as np
```

A scatterplot can be used to plot one dot for each observation. Unlike a line graph, each data point needs an x and y value. Scatterplots are quite useful in visualizing and identifying trends and correlations in data.

Let's say for instance we want to track the league batting average trend over the past 10 years. 

```python
year = [2013,2014,2015,2016,2017,2018,2019,2020,2021,2022]
batting_average = [.253,.251,.254,.255,.255,.248,.252,.245,.244,.243] 

plt.scatter(year, batting_average)
plt.show()
```

### Add a trendline visual

For now we will keep it simple and just add a trendline visual. We will get into more technical ways of presenting this in the future.

```python
year = [2013,2014,2015,2016,2017,2018,2019,2020,2021,2022]
batting_average = [.253,.251,.254,.255,.255,.248,.252,.245,.244,.243] 

plt.scatter(year, batting_average)

z = np.polyfit(year, batting_average, 1)
p = np.poly1d(z)
plt.plot(year,p(year),"r--")

plt.show()
```

# Bar Graphs

Now what if we want an old classic, the bar graph?

We are comparing receptions on the top 5 pass-catchers of a team for a given year. A bar graph is a nice and easy way to visualize this.

```python
#2022 New England Patriots

players = ['Rhamondre Stevenson','Jakobi Meyers','Hunter Henry','Kendrick Bourne','DaVante Parker']
receptions = [69,67,41,35,31]

plt.bar(players,receptions)
plt.show()
```

We can flip this same barchat horizonatally just by replacing bar() with barh().

```python
#2022 New England Patriots

players = ['Rhamondre Stevenson','Jakobi Meyers','Hunter Henry','Kendrick Bourne','DaVante Parker']
receptions = [69,67,41,35,31]

plt.barh(players,receptions)
plt.show()
```

### A few style add-ons

```python
#2022 New England Patriots

players = ['Rhamondre Stevenson','Jakobi Meyers','Hunter Henry','Kendrick Bourne','DaVante Parker']
receptions = [69,67,41,35,31]

plt.barh(players,receptions, color = "hotpink")
plt.title("New England Patriots 2022 Reception Leaders")
plt.show()
```

As you can imagine, we can get a quite bit more creative with these so let your imagination run wild!
