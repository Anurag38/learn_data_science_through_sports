# Matplotlob Plotting

Matplotlib is one of the most widely used Python visualization modules. It can be used to create a variety of different plots, charts, and visualizations. When being introduced to, or even just taking a deeper dive into a module, I ALWAYS encourage my students to get familiar with the documentation. Here is Matplotlib's official documentation page:https://matplotlib.org/stable/index.html

Let's get used to the "import module" statement whenever we are using a module. Additionally, the second line here shows how we import a "submodule". pyplot is a prominent submodule of matplotlib and we import it using the syntax on the second line below.

```python
import matplotlib
from matplotlib import pyplot as plt
```

Let's start with the default plot from the pyplot submodule. A line graph. Line graphs are used for so many applications in sports. Let's show how we can generate some simple ones.

First off, a visual showing points scored by a team in its last 10 games.

```python
points_scored = [102,122,99,87,105,132,98,104,90,123]

plt.plot(points_scored)
plt.show()
```

If you do not specify a list or array for the x-axis, the default will be the length of list/array of the y-axis (0-10 here).

Just looking at this, the viewer cannot really tell what's going on here. This is quite ugly and uninformative. Let's make this a little more appealing.

### Markers

On a line graph, it is sometimes easier to ingest the presented information when each data point along the line is emphasized. Let's start with our first keyword argument in pyplot, "marker".

```python
points_scored = [102,122,99,87,105,132,98,104,90,123]

plt.plot(points_scored, marker= 'o')
plt.show()
```

### Change how the line looks

```python
points_scored = [102,122,99,87,105,132,98,104,90,123]

plt.plot(points_scored, '--', marker= 'o')
plt.show()
```

### Change the line color

```python
points_scored = [102,122,99,87,105,132,98,104,90,123]

plt.plot(points_scored, '--', marker= 'o', color='r')
plt.show()
```

### Change the marker color

```python
points_scored = [102,122,99,87,105,132,98,104,90,123]

plt.plot(points_scored, '--', marker= 'o', color='r', mec = 'k', mfc = 'g')
# mec = outer border of marker, mfc = inside of marker
plt.show()
```

Now it looks a little "prettier" now, but still, it's very hard to understand what is happening here. And what if we had another's teams's stats?

### Plot Multiple Lines

```python
points_scored1 = [102,122,99,87,105,132,98,104,90,123]
points_scored2 = [100,112,90,140,99,127,101,109,90,88]

plt.plot(points_scored1, '--', marker= 'o', color='r', mec = 'k', mfc = 'g')
plt.plot(points_scored2, '--', marker= '*', color='b', mec = 'm', mfc = 'g')

plt.show()
```

### Add Axis Lables and a Title

```python
points_scored1 = [102,122,99,87,105,132,98,104,90,123]
points_scored2 = [100,112,90,140,99,127,101,109,90,88]

plt.plot(points_scored1, '--', marker= 'o', color='r', mec = 'k', mfc = 'g')
plt.plot(points_scored2, '--', marker= '*', color='b', mec = 'm', mfc = 'g')

plt.title("PPG by Team Last 10 Games")
plt.xlabel("Game Number")
plt.ylabel("Team Points")

plt.show()
```

Let's tell this story a little better by labeling the lines to team names and adding a legend.

```python
points_scored1 = [102,122,99,87,105,132,98,104,90,123]
points_scored2 = [100,112,90,140,99,127,101,109,90,88]

plt.plot(points_scored1, '--', marker= 'o', color='r', mec = 'k', mfc = 'g',label="Lakers")
plt.plot(points_scored2, '--', marker= '*', color='b', mec = 'm', mfc = 'g',label="Hawks")

plt.title("PPG by Team Last 10 Games")
plt.xlabel("Game Number")
plt.ylabel("Team Points")
plt.legend(loc="upper left")


plt.show()
```

## Show multiple plots on one graphic

Say we have those PPG stats, but we also want to see each team's APG and RPG over the last 10 games for a much more telling comparison.

The subplot() function! subplot(x,y,z) where x = # of rows, y = # of columns, z = # plot

```python
#Points
points_scored1 = [102,122,99,87,105,132,98,104,90,123]
points_scored2 = [100,112,90,140,99,127,101,109,90,88]

plt.subplot(3,1,1)
plt.plot(points_scored1, '--', marker= 'o', color='r', mec = 'k', mfc = 'g',label="Lakers")
plt.plot(points_scored2, '--', marker= '*', color='b', mec = 'm', mfc = 'g',label="Hawks")

plt.title("PPG by Team Last 10 Games")
plt.xlabel("Game Number")
plt.ylabel("Team Points")
plt.legend(loc="upper left")

#Assists
assists1 = [23,30,19,26,23,35,31,34,24,20]
assists2 = [27,33,22,37,21,22,28,30,21,23]

plt.subplot(3,1,2)
plt.plot(assists1, '--', marker= 'o', color='r', mec = 'k', mfc = 'g',label="Lakers")
plt.plot(assists2, '--', marker= '*', color='b', mec = 'm', mfc = 'g',label="Hawks")

plt.title("APG by Team Last 10 Games")
plt.xlabel("Game Number")
plt.ylabel("Team Assists")
plt.legend(loc="upper left")

#Rebounds
rebounds1 = [43, 50, 39, 46, 43, 55, 51, 54, 44, 40]
rebounds2 = [48, 54, 43, 58, 42, 43, 49, 51, 42, 44]

plt.subplot(3,1,3)
plt.plot(rebounds1, '--', marker= 'o', color='r', mec = 'k', mfc = 'g',label="Lakers")
plt.plot(rebounds2, '--', marker= '*', color='b', mec = 'm', mfc = 'g',label="Hawks")

plt.title("RPG by Team Last 10 Games")
plt.xlabel("Game Number")
plt.ylabel("Team Rebounds")
plt.legend(loc="upper left")


plt.show()
```

This tells a nicer story that what we have so far. But definitely a little jumbled and hard to read. Let's make this presentation worthy!

Let's the whole figure larger and remove each individual x-axis label as they are redundant. The titles of each individual figure are also a little overkill given that the y-axis already tells what the lines represent. So let's get rid of those and add a "Supertitle".

```python
plt.figure(figsize=(12, 8)) #Set overall figure size

#Points
points_scored1 = [102,122,99,87,105,132,98,104,90,123]
points_scored2 = [100,112,90,140,99,127,101,109,90,88]

plt.subplot(3,1,1)
plt.plot(points_scored1, '--', marker= 'o', color='r', mec = 'k', mfc = 'g',label="Lakers")
plt.plot(points_scored2, '--', marker= '*', color='b', mec = 'm', mfc = 'g',label="Hawks")

# plt.title("PPG by Team Last 10 Games")
# plt.xlabel("Game Number")
plt.ylabel("Team Points")
plt.legend(loc="upper left")

#Assists
assists1 = [23,30,19,26,23,35,31,34,24,20]
assists2 = [27,33,22,37,21,22,28,30,21,23]

plt.subplot(3,1,2)
plt.plot(assists1, '--', marker= 'o', color='r', mec = 'k', mfc = 'g',label="Lakers")
plt.plot(assists2, '--', marker= '*', color='b', mec = 'm', mfc = 'g',label="Hawks")

# plt.title("APG by Team Last 10 Games")
# plt.xlabel("Game Number")
plt.ylabel("Team Assists")
plt.legend(loc="upper left")

#Rebounds
rebounds1 = [43, 50, 39, 46, 43, 55, 51, 54, 44, 40]
rebounds2 = [48, 54, 60, 58, 42, 43, 53, 51, 50, 44]

plt.subplot(3,1,3)
plt.plot(rebounds1, '--', marker= 'o', color='r', mec = 'k', mfc = 'g',label="Lakers")
plt.plot(rebounds2, '--', marker= '*', color='b', mec = 'm', mfc = 'g',label="Hawks")

# plt.title("RPG by Team Last 10 Games")
plt.xlabel("Game Number")
plt.ylabel("Team Rebounds")
plt.legend(loc="upper left")

plt.suptitle("Lakers/Hawks Per Game Stats L10")

plt.show()
```
