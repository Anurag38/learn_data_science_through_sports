# Histograms

Histograms show frequency distributuions of a dataset. This is useful for identifying types of distributions and features of a distribution in a concise way.

A very simple example of a histogram is to observe team wins at the end of the season for the entire NFL. This is a good way to examine "parity" in the league.

```python
#2022 NFL Season
win_totals_by_team = [14,14,13,13,13,12,12,10,10,9,9,9,9,9,9,8,8,8,8,7,7,7,7,7,7,6,5,5,4,4,3,3]
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

Simple Histogram

```python
plt.hist(win_totals_by_team, bins=17,rwidth=10)
plt.show()
```

Histograms tend to be more useful with larger sets of data which we will get to.

### Add some useful style

```python
plt.hist(win_totals_by_team, bins=17,rwidth=10)
plt.title("Win Totals by number of Teams")
plt.xlabel("# of Wins")
plt.ylabel("# of Teams")

plt.show()
```

For now, just be familiar with how a histogram is generated with matplotlib.

# Pie Charts

Pie charts are another way to visualize data mainly percentages of a whole. We have a team's PPG scored by each of their starters and we want to see how "balanced" our team's scoring is.

```python
#2022-2023 Los Angeles Clippers
players = ['Paul George','Kawhi Leonard','Norman Powell','Marcus Morris','John Wall']
ppg = [23.3,22.0,16.8,13.0,11.4]

plt.pie(ppg)
plt.show()
```

This doesn't tell us much, so let's add the player labels and a title!

```python
#2022-2023 Los Angeles Clippers
players = ['Paul George','Kawhi Leonard','Norman Powell','Marcus Morris','John Wall']
ppg = [23.3,22.0,16.8,13.0,11.4]

plt.pie(ppg, labels = players)
plt.title("PPG of Clippers Starters 2022-23")
plt.show()
```
