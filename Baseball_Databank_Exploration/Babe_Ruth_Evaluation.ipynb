## Let's actually start analyzing!

Let's take some first steps and do a pretty basic analysis. First, we need to "feature engineer" some well known baseball stats into our dataset. 

#### Batting Average

The age old metric of success. Simply dividing a player's hits by their at-bats.

```python
batting_table['BA'] = round(batting_table['H']/batting_table['AB'],3)
batting_table
```

### Let's add in a few more

Slugging percentage: total bases divided by at-bats. (1B+2B * 2+3B * 3+HR * 4)/AB

Wait! We don't have singles in this dataset. However, given what we know about baseball, Singles = Hits - (2B+3B+HR)

```python
batting_table['Singles'] = batting_table['H'] -   (batting_table['2B'] + batting_table['3B'] + batting_table['HR'])
batting_table
```

Now for slugging percentage.

```python
batting_table['SLG'] = round((batting_table['Singles'] + 2*batting_table['2B'] + 3*batting_table['3B'] + 4*batting_table['HR'])/batting_table['AB'],3)
batting_table
```

On Base Percentage

OBP = (Hits + Walks + Hit by Pitch)/(At Bats + Walks + Hit by Pitch + Sac Flies)

```python
batting_table['OBP'] = round((batting_table['H']+batting_table['BB']+batting_table['HBP'])/(batting_table['AB']+batting_table['BB']+batting_table['HBP']+batting_table['SF']),3)
batting_table
```

I am going to cut it off here for now. As we know, there are a ton more advanced stats and metrics in baseball. For this example, I wanted to show you how to "create" new features in a datset to be used for analysis.

## Compare Babe in his "era"

#### What years did he play?

```python
babes_career = batting_table.loc[batting_table['playerID']== babe_ruth_ID]
start = babes_career['yearID'].min()
finish = babes_career['yearID'].max()
print(start,"-",finish)
```

So let's start with just taking those years for now.

### Parts of a Whole

I think a fun way to look at a player's dominance that is "era-agnostic" is to simply see how much of a given statistic they account for. This encompasses a lot of information and is easy to quantify.

#### Home Runs

As we have established, Babe Ruth played from 1914 to 1935. Let's see how his % percentage of home runs looked like over these years.

```python
home_runs = batting_during_babes_career[['playerID','yearID','HR']]
home_runs.head()
```

```python
#Isolate Babe Ruth's Home Runs and also Summing Home Runs for the entire league by year
babe_home_runs = home_runs[['yearID','HR']].loc[home_runs['playerID'] == babe_ruth_ID]
total_home_runs = home_runs.groupby('yearID', as_index=False).sum()
```

Merging these two tables together on yearID and doing Babe Ruth's Home Runs / Total League Home Runs *100

```python
percent_babe = pd.merge(babe_home_runs,total_home_runs, left_on='yearID', right_on='yearID')
percent_babe['PercentHRs'] = percent_babe['HR_x']/percent_babe['HR_y']*100
percent_babe.head()
```

```python
plt.plot(percent_babe['yearID'],percent_babe['PercentHRs'])
plt.title('Babe Ruth % of League Total HRs')
plt.xlabel('Year')
plt.ylabel('% of Total')
plt.show()
```

#### Percentage of Home Runs throughout his entire career

```python
total_home_runs = home_runs['HR'].sum()
total_babe_home_runs = babe_home_runs['HR'].sum()

round(total_babe_home_runs/total_home_runs*100,2)
```

So over the couse of Babe Ruth's entire career, he accounts for a staggering 3.51% of all home runs hit!

## Our friend, the function

First, let's call back in our playerIDs.

```python
home_run_king_IDs = pd.merge(hrab_df,batting_table, left_on=['nameFirst','nameLast'], right_on=['nameFirst','nameLast'] )
home_run_king_IDs = set(home_run_king_IDs['playerID'])
home_run_king_IDs = list(home_run_king_IDs)
home_run_king_IDs
```

It is usually wise to map out the logic of a function (at least mentally) before writing it. But let's essentially go over the steps of how we calculated HR % for Babe Ruth in his given era.
1. Define Years of Player's Career
2. Re-define battle table to only include stats from player's "era"
3. Define Home Runs Table
4. Isolate Player's Home Runs
5. Summate League's Home Runs
6. Player's Home Runs/League's Home Runs*100
7. Round and Store!

```python
#Instantiate Empty List
home_run_kings_list = []

def percent_HR(playerID):
    
    players_career = batting_table.loc[batting_table['playerID']== playerID]
    start = players_career['yearID'].min()
    finish = players_career['yearID'].max()
    
    batting_during_players_career = batting_table.loc[(batting_table['yearID'] >= start) & (batting_table['yearID'] <= finish)]
    
    home_runs = batting_during_players_career[['playerID','yearID','HR']]
    
    players_home_runs = home_runs[['yearID','HR']].loc[home_runs['playerID'] == playerID]
    total_home_runs = home_runs.groupby('yearID', as_index=False).sum()
    
    total_home_runs = home_runs['HR'].sum()
    total_player_home_runs = players_home_runs['HR'].sum()
    
    percent_HR = round(total_player_home_runs/total_home_runs*100,2)
    
    entry = [playerID, percent_HR]
    
    home_run_kings_list.append(entry)
    
    
    return 
```

Now let's loop it over the list of our home run king players!

```python
for x in home_run_king_IDs:
    percent_HR(x)
```

Let's take this list and turn it into a DataFrame!

```python
home_run_kings_percentage_df = pd.DataFrame(home_run_kings_list, columns = ['playerID','percentHR'])
home_run_kings_percentage_df
```

Bring back in the player's names for a prettier visual.

```python
home_run_kings_percentage_df = pd.merge(home_run_kings_percentage_df, master_table, left_on='playerID',right_on='playerID')
```

### Plot It!

```python
players = (home_run_kings_percentage_df['nameFirst'] + ' ' + home_run_kings_percentage_df['nameLast']).values
hr_percent = home_run_kings_percentage_df['percentHR'].values
```

```python
plt.barh(players, hr_percent)
plt.title("Home Run Percentage of Total League HRs During Player's Career")
plt.xlabel("Percentage of Home Runs")
plt.show()
```
